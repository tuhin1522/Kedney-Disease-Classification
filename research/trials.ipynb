### Download data from drive

```python
import gdown

url = "https://drive.google.com/file/d/1WSK-5rF2zvIf3x7YudZqlH5pegDIJAwF/view?usp=sharing"
```

```python
file_id = url.split("/")[-2]
file_id
```

```python
prefix = "https://drive.google.com/uc?export=download&id="
gdown.download(prefix + file_id, "kidney-CT-Scan-data.zip")
```
